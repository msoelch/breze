Stochastic Recurrent Networks for human movement modelling
===============================================================================

This notebook presents how to use stochastic recurrent networks (STORNs) for 
modelling robot movement. The method has been proposed in 

 * Bayer, Justin, and Christian Osendorfer. "Learning Stochastic Recurrent Networks." arXiv preprint arXiv:1411.7610 (2014). [Link](arxiv.org/abs/1411.7610).
 
It is best to revisit the paper to get a theoretical understanding of the method,
but a quick recap follows.

The model
---

We are interested in modelling sequences $\{x_{1:T}\}_{i=1}^N$ over a vectorial
space. To do so, we assume that the data is generated by the following model:

$$h_t = f(h_{t-1}W + x_t V + z_t \hat{V} + b), \\
x_{t+1} = g(h_t Q + \hat{b}).$$

Here, $z_t \sim \mathcal{N}(0, 1)$ are **stochastic units** which help to model
the data.

As can be seen, the model is a generalisation of plain RNNs: if we set 
$\hat{V} = \mathbf{0}$, we will recover an Elman network.
However, adding stochastic units can greatly enhance the predictive capabilities 
of our model. 

For one, it lets us model dependencies between the components of each $x_t$.
Further, it can express more complicated distributions. For example, the output
of an RNN is just a point. If we do not attach any probabilistic meanings to
those points, they will form a Dirac distribution. However, if we integrate all
the $z_t$ out, we can get any (yes, any) distribution--Gaussian, Laplace, 
student's T or even mixtures.


The learning algorithm/estimator
---

We are using a certain flavor of variational inference, namely *stochastic gradient variational Bayes* (SGVB)
to find good paramters--that is $\{h_0, V, \hat{V}, Q, b, \hat{b}\}$--for our model.
SGVB has been described in 

 * Kingma, Diederik P., and Max Welling. "Auto-encoding variational bayes." arXiv preprint arXiv:1312.6114 (2013) and
 * Rezende, Danilo Jimenez, Shakir Mohamed, and Daan Wierstra. "Stochastic backpropagation and approximate inference in deep generative models." arXiv preprint arXiv:1401.4082 (2014).
 
but the STORN article features a quick recap. We will do a shorter one.

The main idea is to minimize an upper bound on the negative log-likelihood, 
which also makes a variational approximation $q(z|x)$ close to the true posterior 
$p(z|x)$. This is done by making $q$ a parameterised model conditioned on the
input (which is weird in the first moment); in the original papers, a neural net
was used for that. Here, we are using (bidirectional) recurrent networks. The
parameters of that model (the recognition model or inference model) are optimised
jointly with the generative model above.

Doing so, we get something for free: $q$ can be used for all kinds of approximate
inference. We can perform an approximate denoising by looking at an observation $x$,
obtaining the maximum of the approximate posterior $\text{arg max}_{\hat{z}} q(\hat{z}|x)$
and then reconstruct $\hat{x}$ from that.

We can also do impute missing values by doing so many times (using sampling from $q$ 
instead of MAP) and re-feed each $\hat{x}$ into our model. This is equivalent to MCMC
sampling from the model. If we clamp some components of $x$, we will do proper missing
value imputation.

Furthermore, there exists an efficient estimator for the NLL that does not rely on
a costly Markov chain--something which is not available for the RBM family.

Data format
-------------------------------------------------------------------------------

A word on the data format. To make things efficient, we need to put all the 
sequential data in 3D arrays. This is done by padding the shorter sequences
with zeros. To make sure that those zeros don't interfer with learning, we will
also use masks (that is binary matrices) with which the component wise loss
is multiplied--that way, only the real data points will contribute to it.

It will also mean longer processing times as if all sequences were of the same
length. But it is way faster than working through each one separately.

An array is thus of the form

  ``(T, N, D)``
  
where ``T`` is the number of time steps, ``N`` is the number of samples and 
``D`` is the dimensionality of the data.

The training and validation data will also be chunked into smaller sequences,
e.g. ``100``. This is not done for the testing data. I personally do not 
believe that it affects performance much.

Naming convention for data
---

Throughout Breze, ``X`` is the training data, ``VX`` is the validation data
and ``TX`` is the testing data. ``M``, ``VM`` and ``TM`` are the corresponding
masks.

Often, we will not chunk the testing data. In that case, the testing data is
contained in a list called ``txs``, which contains many sequences of differing
length of the shape ``(T, D)``. In this case, we only have a single testing
sequence, which is why it is called ``tx``.

```python
# This is taken from Graham Taylor--reading in the human motion data.
# https://gist.github.com/gwtaylor/2505670

def preprocess_data(Motion):
 
    n_seq = Motion.shape[1]
 
    # assume data is MIT format for now
    indx = np.r_[
        np.arange(0,6),
        np.arange(6,9),
        13,
        np.arange(18,21),
        25,
        np.arange(30,33),
        37,
        np.arange(42,45),
        49,
        np.arange(54,57),
        np.arange(60,63),
        np.arange(66,69),
        np.arange(72,75),
        np.arange(78,81),
        np.arange(84,87),
        np.arange(90,93),
        np.arange(96,99),
        np.arange(102,105)]
 
    row1 = Motion[0,0][0]
 
    offsets =   np.r_[
        row1[None,9:12],
        row1[None,15:18],
        row1[None,21:24],
        row1[None,27:30],
        row1[None,33:36],
        row1[None,39:42],
        row1[None,45:48],
        row1[None,51:54],
        row1[None,57:60],
        row1[None,63:66],
        row1[None,69:72],
        row1[None,75:78],
        row1[None,81:84],
        row1[None,87:90],
        row1[None,93:96],
        row1[None,99:102],
        row1[None,105:108]]
 
    # collapse sequences
    batchdata = np.concatenate([m[:, indx] for m in Motion.flat], axis=0)
 
    data_mean = batchdata.mean(axis=0)
    data_std = batchdata.std(axis=0)
 
    batchdata = (batchdata - data_mean) / data_std
 
    # get sequence lengths
    seqlen = [s.shape[0] for s in Motion.flat]
 
 
    return batchdata, seqlen, data_mean, data_std
 
def load_data_rather_raw(filename):
    # load data post preprocess1
    mat_dict = loadmat(filename)
    Motion = mat_dict['Motion']
 
    batchdata, seqlen, data_mean, data_std = preprocess_data(Motion)
 
    # put data into shared memory
    x = np.asarray(batchdata, dtype=theano.config.floatX)
 
    return x, seqlen, data_mean, data_std


def load_data(x, seqlen, T=200):
    xs = [x[seqlen[0] + seqlen[1]:]]
    vxs = [x[seqlen[0]:seqlen[0] + seqlen[1]]]
    tx = x[:seqlen[0]]
    
    x = split(xs, T)
    vx = split(vxs, T)
    
    X, M = padzeros(x, False, True)
    VX, VM = padzeros(vx, False, True)
    
    X, M, VX, VM = [interleave(i) for i in (X, M, VX, VM)]
    
    print M.shape
    
    M = M.mean(2)[:, :, np.newaxis]
    VM = VM.mean(2)[:, :, np.newaxis]

    print 'size of X', X.size
    print 'size of VX', VX.size

    X, VX, M, VM = [base.cast_array_to_local_type(i) for i in (X, VX, M, VM)]
    tx = base.cast_array_to_local_type(tx)

    print 'training shape', X.shape, M.shape
    print 'validation shape', VX.shape, VM.shape
    print 'testing shape', tx.shape

    return X, M, VX, VM, [tx]

x, seqlen, mean, std = load_data_rather_raw('motion.mat')
X, M, VX, VM, tx = load_data(x, seqlen, 100)

print 'Masking ratios: train %g val %g' %(M.mean(), VM.mean())
print '(That is the fraction of elements in the data arrays which do not stem from padding.)'
```

```python
def load_data(datafile, T=50):
    with h5py.File(datafile) as fp:
        dev_seqs = [np.array(fp['dev'][i]) for  i in fp['dev']]
        txs = [np.array(fp['test'][i]) for  i in fp['test']]
        tzs = np.array(fp['test_anomaly_label'])

    x = split(dev_seqs[:80], T)
    m = np.concatenate(x).mean(axis=0)
    s = np.concatenate(x).std(axis=0)
    
    x = [(xx-m)/s for xx in x]
    vx = [(vxx-m)/s for vxx in split(dev_seqs[80:], T)]
    txs = [(tx-m)/s for tx in txs]
    
        
    X, M = tuple(interleave(l) for l in padzeros(x, front=False, return_mask=True))
    VX, VM = tuple(interleave(l) for l in padzeros(vx, front=False, return_mask=True))
    TX, TM = tuple(interleave(l) for l in padzeros(txs, front=False, return_mask=True))
 
    X, VX = [base.cast_array_to_local_type(i) for i in (X, VX)]
    txs = [base.cast_array_to_local_type(i) for i in txs]
    
    print 'training shape', X.shape, M[:,:,:1].shape
    print 'validation shape', VX.shape, VM[:,:,:1].shape
    print 'test shape', TX.shape, TM[:,:,:1].shape, tzs.shape
    
    return X, M[:,:,:1], VX, VM[:,:,:1], txs, tzs#TX, TM[:,:,:1], tzs



X, M, VX, VM, txs, tzs = load_data('P:/Datasets/BaxterCollision/data/bluebear/bluebear.h5')
```

Model initialisation
--------------------------------------------------------------------------------

Here we initialise the model. We will comment the most important parameters in the
cell. The hyper parameters were obtained using Bayesian optimisation via whetlab.

```python
# Adadelta with Nesterov momentum--powerful optimiser for FD-RNNs.

optimizer = 'adadelta', {'step_rate': 1, 
                        'momentum': 0.99, 
                        'decay': 0.78, 
                        'offset': 0.00005, 
                        'step_adapt': False, 'step_rate_min': 1e-8, 'step_rate_max': 1}
# Batch sizes speed up learning for RNNs as well.
batch_size = 5

#  Amount of hidden layers of the recognition and generating model.
n_gen_layers = 1
n_recog_layers = 1

n_hiddens_recog = [50] * n_recog_layers
transfers_recog = ['tanh'] * n_recog_layers
n_hiddens_gen = [50] * n_gen_layers
transfers_gen = ['tanh'] * n_gen_layers
n_latents = 8

# Standard deviation for a Normal from which all parameters are drawn.
par_std = 0.5

# The input-to-hidden weights are drawn from this instead.
in_to_hidden_par_std = 0.000001

# After initialisation, the spectral radius of the recurrent matrices is set to this value.
spectral_radius = 4.65

# The recurrent weight matrices are restricted to only have 50 non-zero elements per row.
sparsify_rec = 50

# Similar for the input matrix.
sparsify_in = 2

# Fast dropout rates for the *generating* model.
p_dropout_inpt = .5
p_dropout_hiddens = .5
p_dropout_hidden_to_out = .12
p_dropout_shortcut = .35

reload(sgvb)

# The assumptions we make. First, we assume that the noise level 
# of the visibles is constant and somewhere around .1 everywhere.
# We assume the latent variables to follow a white Gaussian.
class Assumptions(sgvb.ConstantVarVisibleGaussAssumption, sgvb.DiagGaussLatentAssumption):
    out_std = theano.shared(np.array(.1).astype(theano.config.floatX))

assumptions = Assumptions()

# We can use a bidirectional or unidirectional recognition model. 
# Not possible for the generating model.
#m = sgvb.StochasticRnn(
m = sgvb.BidirectStochasticRnn(
    X.shape[2],
    n_hiddens_recog, n_latents, n_hiddens_gen, transfers_recog, transfers_gen, 
    use_imp_weight=True,
    assumptions=assumptions,
    batch_size=batch_size,
    optimizer=optimizer,
    p_dropout_inpt=p_dropout_inpt,
    p_dropout_hiddens=p_dropout_hiddens,
    p_dropout_hidden_to_out=p_dropout_hidden_to_out,
    p_dropout_shortcut=p_dropout_shortcut)

infos = []
m.exprs['true_loss'] = m.exprs['loss']

m.gradient_clip_threshold = None
```

```python
m.initialize(
    par_std=par_std, 
    par_std_in=in_to_hidden_par_std, 
    sparsify_affine=sparsify_in, 
    sparsify_rec=sparsify_rec, 
    spectral_radius=None)#spectral_radius)
```

```python
sample_map = m.vae.latent[:, :, :n_latents]
gen_inpt = T.concatenate([sample_map, m.vae.condition], axis=2)

map_recons_exprs = theano.clone(m.exprs['output'], 
                                {m.vae.sample: sample_map})
f_map_recons = m.function(['inpt'], map_recons_exprs)
```

```python
def map_impute(start, to_impute_idxs, n_steps=10):
    denoised = start.copy()
    denoised[to_impute_idxs] = np.random.standard_normal(denoised[to_impute_idxs].shape)
    noisy = denoised.copy()
    
    for i in range(n_steps):
        reconstructed = f_map_recons(denoised)
        denoised[to_impute_idxs] = reconstructed[:, :, :49][to_impute_idxs]
    return noisy, denoised
```
